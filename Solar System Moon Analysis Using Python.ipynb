### Importing Libraries

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

#### Import Moon Data

```python
# Importing and Retrieving the table consisting data of moon
moon_data_df = pd.read_html('https://en.wikipedia.org/wiki/List_of_natural_satellites#List', match='Parent')[0]
```

#### Displaying sample data

Displaying first 10 rows to perform initial analysis about the moon data.

```python
# display first 10 rows
moon_data_df.head(10)
```

#### Renaming columns

Renaming columns so that it increases the familiarity and interpretability of data by someone unfamiliar to the data easier.

Renaming the columns should make the data easy to read and clean. Replacing 'Parent' with 'Parent of the Moon' makes it clear about the moon's parent planet. Renaming 'Numeral' to 'Roman Numeral' justifies the observation it consists of, also replacing discovery with identification and adding moon to the column names makes it easier for a non-scientific background person to understand.

```python
# renaming columns using map
moon_data_rename_map = {'Parent' : 'Parent of Moon', 'Numeral' : 'Roman Numeral', 'Name':'Name of the Moon', 'Discovery year' : 'Moon Discovery Year', 'Orbital semi-major axis (km)' : 'Largest radius in orbit (km)', 'Sidereal period (d) (r = retrograde)' : 'Time required for one revolution w.r.t. parent (d) (r = retrograde)'}
moon_data_df.rename(columns = moon_data_rename_map, inplace = True)
```

```python
# display output
moon_data_df.head(10)
```

#### Reordering data for better understandability

Rearranging the data so that the most significant observations come first.

Since we are studying the data to recognize the dwarf planets of the solar system, it is useful to get basic details about a planet before using the specific scientific information about the planet.

Parent of the moon is in the first column since it assists with the identification of the dwarf planet, followed by the moon's name and its roman numeral. Other information about the planet is followed by the former since they require in-depth knowledge to be understood.

```python
# reordering columns and display them
reordered_moon_data_df = moon_data_df[['Parent of Moon', 'Name of the Moon', 'Roman Numeral', 'Moon Discovery Year', 'Discovered by', 'Mean radius (km)', 'Largest radius in orbit (km)', 'Time required for one revolution w.r.t. parent (d) (r = retrograde)', 'Notes', 'Ref(s)', 'Image']]
reordered_moon_data_df.head(10)
```

### Data Wrangling

Data wrangling or data cleaning refers to a variety of processes designed to transform raw data into more readily used formats.

Here, we will be cleaning the data using basic techniques to remove inherent errors and make it distort free to perform analysis and gain insights about the data.

##### Step 1 - Splitting data into multiple columns for easier understandability

We will be splitting Romon numeral into two columns named Roman and Moon number since it consists of Roman numeric data with Numeric (integer). This step is necessary to make analysis about the relation between parent plane and their corresponding number of moons understandable and interpretable.

```python
# splitting composite column
reordered_moon_data_df[['Roman','Moon number']] = reordered_moon_data_df['Roman Numeral'].str.split(" ",expand=True)
```

Now, we will be extracting integer Moon number from the column using Regular Expression.

```python
# extracting numeric data using regex
reordered_moon_data_df['Moon number'] = reordered_moon_data_df['Moon number'].str.extract('(\d+)')
```

Now we will be rearraning columns to further our analysis.

```python
# rearrange columns
reordered_moon_data_df = reordered_moon_data_df[['Parent of Moon', 'Moon number', 'Name of the Moon', 'Moon Discovery Year', 'Discovered by', 'Roman', 'Mean radius (km)', 'Largest radius in orbit (km)', 'Time required for one revolution w.r.t. parent (d) (r = retrograde)', 'Notes', 'Ref(s)', 'Image', 'Roman Numeral']]
reordered_moon_data_df
```

##### Step 2 - Cleaning and approximation of data

We are cleaning the scientific numerical data by removing various symbols like '>', '<', '~' so that it is easier to perform calculations and analysis on the numeric data.

The process involves cleaning all the columns consisting of such data and later converting them to numeric data type.

```python
# cleaning 'Largest radius in orbit (km)'
reordered_moon_data_df['Largest radius in orbit (km)'] = reordered_moon_data_df['Largest radius in orbit (km)'].apply(lambda s : s.replace('>', ''))
reordered_moon_data_df['Largest radius in orbit (km)'] = reordered_moon_data_df['Largest radius in orbit (km)'].apply(lambda s : s.replace(',', ''))
reordered_moon_data_df['Largest radius in orbit (km)'] = reordered_moon_data_df['Largest radius in orbit (km)'].apply(lambda s : s.split('±')[0])
reordered_moon_data_df['Largest radius in orbit (km)'] = reordered_moon_data_df['Largest radius in orbit (km)'].apply(lambda s : s.replace('(unknown)', ''))
```

```python
# cleaning 'Mean radius (km)'
reordered_moon_data_df['Mean radius (km)'] = reordered_moon_data_df['Mean radius (km)'].apply(lambda s : s.replace('>', ''))
reordered_moon_data_df['Mean radius (km)'] = reordered_moon_data_df['Mean radius (km)'].apply(lambda s : s.replace('<', ''))
reordered_moon_data_df['Mean radius (km)'] = reordered_moon_data_df['Mean radius (km)'].apply(lambda s : s.replace('≈', ''))
reordered_moon_data_df['Mean radius (km)'] = reordered_moon_data_df['Mean radius (km)'].str.extract('(\d*[.]{0,1}\d*)', flags = 0)
```

```python
# cleaning 'Time required for one revolution w.r.t. parent (d) (r = retrograde)'
reordered_moon_data_df['Time required for one revolution w.r.t. parent (d) (r = retrograde)'] = reordered_moon_data_df['Time required for one revolution w.r.t. parent (d) (r = retrograde)'].apply(lambda s : s.replace('(unknown)', ''))
reordered_moon_data_df['Time required for one revolution w.r.t. parent (d) (r = retrograde)'] = reordered_moon_data_df['Time required for one revolution w.r.t. parent (d) (r = retrograde)'].apply(lambda s : s.replace('>', ''))
reordered_moon_data_df['Time required for one revolution w.r.t. parent (d) (r = retrograde)'] = reordered_moon_data_df['Time required for one revolution w.r.t. parent (d) (r = retrograde)'].apply(lambda s : s.replace('<', ''))
reordered_moon_data_df['Time required for one revolution w.r.t. parent (d) (r = retrograde)'] = reordered_moon_data_df['Time required for one revolution w.r.t. parent (d) (r = retrograde)'].apply(lambda s : s.replace('≈', ''))
reordered_moon_data_df['Time required for one revolution w.r.t. parent (d) (r = retrograde)'] = reordered_moon_data_df['Time required for one revolution w.r.t. parent (d) (r = retrograde)'].str.extract('(\d*[.]{0,1}\d*)', flags = 0)
```

```python
# display
reordered_moon_data_df.head(10)
```

##### Step 3 - Homogenize the columns

For making data processing and analysis, it is required for all obsrvations of a column to have a single data type.

In the case of 'Moon Discovery Year', a numeric data type.

```python
# change column data type
reordered_moon_data_df['Moon Discovery Year'] = (pd.to_numeric(reordered_moon_data_df['Moon Discovery Year'], errors='coerce'))
reordered_moon_data_df.head(10)
```

##### Step 4 - Modifying the column's data types

For proper analysis and computations about the moon's data, the observations related to scientific research that is based on purely numeric data, there is a need to convert 'Moon number', 'Mean radius (km)' ,'Largest radius in orbit (km)' ,'Time required for one revolution w.r.t. parent (d) (r = retrograde)' to numerical data type.

```python
# change data type to numeric
reordered_moon_data_df['Moon number'] = pd.to_numeric(reordered_moon_data_df['Moon number'])
reordered_moon_data_df['Mean radius (km)'] = pd.to_numeric(reordered_moon_data_df['Mean radius (km)'])
reordered_moon_data_df['Largest radius in orbit (km)'] = pd.to_numeric(reordered_moon_data_df['Largest radius in orbit (km)'])
reordered_moon_data_df['Time required for one revolution w.r.t. parent (d) (r = retrograde)'] = pd.to_numeric(reordered_moon_data_df['Time required for one revolution w.r.t. parent (d) (r = retrograde)'])
```

```python
# display
reordered_moon_data_df.head(10)
```

### Setting index

We have almost done all the Data Wrangling possible. Now the final task is to select a suitable Index for our DataFrame.

Since we are studying about the moons of the planets, therefore it is suitable to use the parent of the moon to be used as primary index since a planet may consists of many moons but a moon may consist of only singular planet. 

And, secondry index will be Moon number by using group by so that primary index is not redundant.

Also, since we are more familiar with the names of the planets than the names of the moon, it makes it easier to read the data about the moons.

The size of the parent planet is much greater than of its child moon therefore making the identification easier to an unfamiliar person.

```python
#setting primary and secondry index
indexed_moon_data_df = reordered_moon_data_df.set_index(['Parent of Moon', 'Moon number']).groupby(level='Parent of Moon')
indexed_moon_data_df.head()
```

## Analysing a research question

We need to analyse if the number of moons of a parent planet is directly proportional to the size of the parent planet?

We will first try to identify the parent planet's size. To do so, we will consider the radius of the moon since greater the radius symbolises greater planet size.

```python
# identifying the the average radius of moons of a planet
moon_avg_radius = indexed_moon_data_df['Largest radius in orbit (km)'].mean()
moon_avg_radius
```

```python
# plotting radius graph
plt.figure(figsize=(20,5))
plt.plot(moon_avg_radius.index, moon_avg_radius.values)
plt.ylabel('Avg radius of moon')
plt.xlabel('Planet')
plt.title("Radius of planet's moons")

plt.show()
```

Now, as we have an estimate about the size of the parent planets. We can count the number of moon per planet.

```python
# count number of moons per planet
num_moons = indexed_moon_data_df['Name of the Moon'].count()
num_moons
```

```python
# plotting number of moon graph
plt.figure(figsize=(20, 3))
plt.bar(num_moons.index, num_moons.values)
plt.title('Count of Moons')
plt.xlabel('Planet')
plt.ylabel('Number of moons')
```

After analysing the above two graphs, we observed the following:
> 1. Jupiter is the biggest planet in our solar system followed by Saturn, Neptune and Uranus.
> 2. Saturn and Jupiter have the highest number of moons.

##### Conclusion

From the above observation, we can conclude that the number of moons of a parent planet is directly proportional to the size of the parent planet since Jupiter, Saturn, Uranus and Neptune have marginally greater number of moons and are comparitively greater in size than other planets.
